```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import os
import glob
```

```python
#testfile="/home/flyranch/2017_10_26_trap_G_raw_data_sample/tl_0006_1882_20161130_102618.jpg"
#image=cv2.imread(testfile)

#directry of all images for making mask
img_dir="/home/flyranch/2017_10_26_trap_G_raw_data_sample"
data_path=os.path.join(img_dir,'*g')
#glob-filename pattern matching
files=glob.glob(data_path)

data=[]
#create list of all images
for file in files:
    img=cv2.imread(file)
    data.append(img)
```

```python
#covert list of images to 4d array
data_array=np.asarray(data)
```

```python
type(data_array)
```

```python
#(n,nrows,ncols,3(color))

data_array[118][1943][2591][2]
```

```python
#get median of all images

med=np.median(data_array,axis=0)
```

```python
type(med)
```

```python
#median of all images(nrows,ncols,3(color))

med[1943][2591][2]
```

```python
#convert to int array because an image with RGB values (0-1 float or 0-255 int)

int_med=med.astype(int)
```

```python
#plt.figure(figsize=[400,400])
plt.imshow(int_med)
plt.show()
```

```python
#write the image file into the folder

filename='/home/flyranch/2017_10_26_trap_G_raw_data_sample/mask1.jpg'
cv2.imwrite(filename,int_med)
```
